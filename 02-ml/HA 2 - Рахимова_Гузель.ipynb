### 4

What is the proportion of passengers who survived the catastrophe? What is the proportion of men and women among the survivors?

```python
# your code here
df = pd.read_csv('train.csv')
df = df[['Survived','Sex','PassengerId']]
surv = df[df['Survived']==1]['Survived'].count() #количесво выживших пассажиров
all =df['Survived'].count()# общее количесво людей на борту
prop_surv = round((surv)/all,2)# доля выживших относительно всех пассажиров округленная до 2х знаков
surv_men = df[(df['Survived']==1) & (df['Sex']== 'male')]['Survived'].count()#количесво выживших мужчин
surv_wom = surv - surv_men #количесво выживших женщин
prop_surv_men=round((surv_men)/surv,2)# доля мужчин среди выживших пассажиров,округленная до 2х знаков
prop_surv_wom=round((surv_wom)/surv,2)# доля жен6щин среди выживших пассажиров,округленная до 2х знаков
print(f'Доля выживших пассажиров {prop_surv}')
print(f'Доля мужчин среди выживших пассажиров {prop_surv_men}')
print(f'Доля женщин среди выживших пассажиров {prop_surv_wom}')
```

***Новая характеристика, разница в цене билета. Я беру медиану цены для определенного места посадки и для определенного класса(получаю примерную стоимость билета, зависящей от места посадки и класса) и высчитываю процент отличия стоимости билета и фактической ценой за билет. Получается отрицательный процент это потеря денег компании, потому что билет был приобретен со скидкой, так как он по стоимости меньше стандартной цены для этого места посадки и класса.
```python
# your code here
df = pd.read_csv('train.csv')
df = df[['Fare','Embarked','Pclass']]

sout1 = df[(df['Embarked'] =='S' ) & (df['Pclass'] == 1)]['Fare'].median()
sout2 = df[(df['Embarked'] =='S' ) & (df['Pclass'] == 2)]['Fare'].median()
sout3 = df[(df['Embarked'] =='S' ) & (df['Pclass'] == 3)]['Fare'].median()

cherb1 = df[(df['Embarked']=='C') & (df['Pclass'] == 1)]['Fare'].median()
cherb2 = df[(df['Embarked']=='C') & (df['Pclass'] == 2)]['Fare'].median()
cherb3 = df[(df['Embarked']=='C') & (df['Pclass'] == 3)]['Fare'].median()

qeen1 = df[(df['Embarked']=='Q') & (df['Pclass'] == 1)]['Fare'].median()
qeen2 = df[(df['Embarked']=='Q') & (df['Pclass'] == 2)]['Fare'].median()
qeen3 = df[(df['Embarked']=='Q') & (df['Pclass'] == 3)]['Fare'].median()

df.loc[(df['Embarked'] =='S' ) & (df['Pclass'] == 1), 'Price_diff'] = round(( df['Fare'] - sout1)*100/sout1,1)
df.loc[(df['Embarked'] =='S' ) & (df['Pclass'] == 2), 'Price_diff'] = round(( df['Fare'] - sout2)*100/sout2,1)
df.loc[(df['Embarked'] =='S' ) & (df['Pclass'] == 3), 'Price_diff'] = round(( df['Fare'] - sout3)*100/sout3,1)

df.loc[(df['Embarked'] =='C' ) & (df['Pclass'] == 1), 'Price_diff'] = round((df['Fare'] - cherb1 )*100/cherb1,1)
df.loc[(df['Embarked'] =='C' ) & (df['Pclass'] == 2), 'Price_diff'] = round((df['Fare'] - cherb2 )*100/cherb2,1)
df.loc[(df['Embarked'] =='C' ) & (df['Pclass'] == 3), 'Price_diff'] = round((df['Fare'] - cherb3 )*100/cherb3,1)

df.loc[(df['Embarked'] =='Q' ) & (df['Pclass'] == 1), 'Price_diff'] = round((df['Fare'] - qeen1 )*100/qeen1,1)
df.loc[(df['Embarked'] =='Q' ) & (df['Pclass'] == 2), 'Price_diff'] = round((df['Fare'] - qeen2 )*100/qeen2,1)
df.loc[(df['Embarked'] =='Q' ) & (df['Pclass'] == 3), 'Price_diff'] = round((df['Fare'] - qeen3 )*100/qeen3,1)
df


```

```python
# your code here
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.preprocessing import StandardScaler


df = pd.read_csv('train.csv')
df_t = pd.read_csv('test.csv')

# убираем не нужные клонки ,заполняем пропуски и преобразуем категориальные строковые признаки
df=df.drop(['Name','Ticket','Cabin','PassengerId'],axis=1)
print('Пропуски в тренировочных данных')
print(df.isnull().sum())

df = df.fillna({'Age' : df['Age'].median(), 'Embarked' : df['Embarked'].mode()[0]})
df = pd.get_dummies(df, columns=['Sex'], prefix='Sex',dtype = int)
df = pd.get_dummies(df, columns=['Embarked'], prefix='Embarked',dtype = int)
print()
print()
print('Пропуски в тренировочных данных после заполнения')
print(df.isnull().sum())

df_t=df_t.drop(['Name','Ticket','Cabin','PassengerId'],axis=1)
print()
print()
print('Пропуски в тестовых данных')
print(df_t.isnull().sum())

df_t = df_t.fillna({'Age' : df_t['Age'].median(), 'Fare' : df_t['Fare'].median()})
df_t = pd.get_dummies(df_t, columns=['Sex'], prefix='Sex',dtype = int)
df_t = pd.get_dummies(df_t, columns=['Embarked'], prefix='Embarked',dtype = int)

print()
print()
print('Пропуски в тестовых данных после заполнения')
print(df_t.isnull().sum())

# делим таблицу на признаки и целевую переменную.
X_train = df.drop(columns='Survived')
y_train = df['Survived']

scaler = StandardScaler()
scaler.fit(X_train)

X_train_normed = scaler.transform(X_train)

knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train_normed, y_train)

preds_train=knn.predict(X_train_normed)

X_test_normed = scaler.transform(df_t)
preds_test = knn.predict(X_test_normed)

df_t = pd.read_csv('test.csv')
predict_test = pd.DataFrame({'PassengerId':df_t['PassengerId'], 'Survived': preds_test})
print()
print()
print('Предсказания на тестовой выборке')
predict_test
```
