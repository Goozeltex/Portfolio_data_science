## Q 15

Alice analyzes her spending in 28 days. Write a program that will help her organize the information and determine the week when the girl spent the most money.

##### INPUT FORMAT
- A text file notes.txt, which records data about the day and expenses on that day. For example, "04.04.2027, restaurants: 600, pharmacies: 200". This line says that on April 4, 2027, Alice spent 600 rubles for the "restaurants" category and 200 rubles for the "pharmacies" category".
- Spending categories can be repeated in one day. The names of the categories in the data can be any, for example, "pharmacies", "trips", "gifts", etc.
- It is guaranteed that the amount of expenses is always an integer.
- It is guaranteed that all dates are in order (from the 1st of the month to the 28th inclusive), that is, there are always exactly four weeks in the data.
Information about expenses for each day is recorded from a new line.


##### OUTPUT FORMAT
- The program should determine the number of the week when Alice spent the most money, and also sort the expenses for that week from the largest to the smallest. 
- The result of the program should be a text file result.txt with strings of the following format:
"The number of the week with the most expenses: <week number>". It is guaranteed that there is only one such week. 
- Next, information about categories and expenses for this week only is displayed in the format "<category name>, <expenses in this category>". Each entry should be output from a new line. Records should be displayed depending on the amount of spending — from the largest to the smallest. It is guaranteed that there are no categories with the same total expenses per week.

##### ADDITIONAL INFO
The open test uses files notes.txt and result.txt. You can download them and test your solution locally.
In the example, the test number is displayed in the first column (0 is an open test), you do not need to read this number or use it in your solution.

### Шаги решения

- Считать и разобрать данные о расходах из файла notes.txt.
- Организовать данные по неделям и категориям.
- Определить неделю с наибольшими расходами.
- Вывести результаты в указанном формате в файл result.txt.

```python
# Шаг 1. Считываем данные из файла

data_list = []

with open('notes(26_15).txt', 'r', encoding='utf-8') as inf:
    for line in inf:
        cats = line.strip().split(', ')[1:]     
        data_list.append(cats)
```

"04.04.2027, restaurants: 600, pharmacies: 200"

Пример строки из файла. Чтобы записать только переменные мы используем срез [1:], чтобы записать все данные о категории об одной строке в одну "строку" (STRING), поэтому каждый элемент в data_list будет списком категорий и трат для одного дня.

```python
# Шаг 2. Организовываем данные по неделям и категориям

all_data = {1: {}, 2: {}, 3: {}, 4: {}}  # словарь, где ключи - номера недель, значения - расходы по категориям

key = 1 # создаем переменную для отсчета недель в цикле. Начинаем с первой недели. 
for i in range(0, len(data_list), 7): # проходим по записям с шагом 7, чтобы получить недельные данные
# i будет принимать значения 0, 7, 14, 21, 28. 
    week = data_list[i:i+7] # срез по дням
    week_data = {} # создаем словарь где будем сохранять сумму расходов по категориям
    for day in week: # проходим по элементам списка week
        for cat in day: # проходим по элементам списка day
            cat, summa = cat.split(': ') # разделяем категорию и сумму
            if cat not in week_data: # если категории нет в словаре, добавляем ее
                week_data[cat] = int(summa) # и записываем сумму
            else: 
                week_data[cat] += int(summa) # если категория уже есть в словаре, прибавляем сумму
    all_data[key] = week_data # записываем данные по неделе в словарь all_data
    key += 1

print(all_data) # посмотрим на словарь
```

```python

# Шаг 3. Определяем неделю с наибольшими расходами

# Теперь нам надо найти неделю с наибольшими расходами.

res = [] # список для сохранения результатов    
for week, w_dict in all_data.items(): # проходим по парам ключ:значение словаря all_data
    res.append((sum(w_dict.values()), week))

print(sorted(res)) # посмотрим на отсортированный список

week_num = sorted(res)[-1][-1] # достаем номер недели с наибольшими расходами 


# Теперь для этой недели надо подготовить траты по категориям

res = []
for k, v in all_data[week_num].items():
    res.append((v, k)) # меняем порядок для сортировки по сумме
```

```python
# Шаг 5. Записываем результаты в файл

with open('result(26_15).txt', 'w', encoding='utf-8') as ouf:
    print(f'Номер недели с наибольшими расходами: {week_num}', file=ouf)
    for cat in sorted(res, reverse=True):
        print(f'{cat[1]}, {cat[0]}', file=ouf)
```

```python
# Полный код

data_list = []

with open('notes.txt', 'r', encoding='utf-8') as inf:
    for line in inf:
        cats = line.strip().split(', ')[1:]     
        data_list.append(cats)

all_data = {1: {}, 2: {}, 3: {}, 4: {}}

key = 1
for i in range(0,len(data_list),7):
    week = data_list[i:i+7]
    week_data = {}
    for day in week:
        for cat in day:
            cat, summa = cat.split(': ')
            if cat not in week_data:
                week_data[cat] = int(summa)
            else:
                week_data[cat] += int(summa)
    all_data[key] = week_data
    key += 1
    
res = []
for week, w_dict in all_data.items():
    res.append((sum(w_dict.values()), week))
week_num = sorted(res)[-1][-1]

res = []
for k, v in all_data[week_num].items():
    res.append((v, k))
    
with open('result.txt', 'w', encoding='utf-8') as ouf:
    print(f'The number of the week with the most expenses: {week_num}', file=ouf)
    for cat in sorted(res, reverse=True):
        print(f'{cat[1]}, {cat[0]}', file=ouf)
```

# Q 14

### Task Description

Ira analyzes the dynamics of the Gini index in African countries. She wants to understand in which countries the index has been steadily growing, and among those – determine the country with the maximum index value for 2019. Write a program that will help her with this task.

#### 📥 INPUT FORMAT
- First, enter the name of the country. Then, from a new line, separated by a comma and a space, enter the values of the Gini index in this country for 2017, 2018 and 2019, respectively. The index values are always real numbers.
- Each line contains information about only one country — either the name or the values of the Gini index.
- It is guaranteed that the names of the countries are not repeated.
- When the recordings end, the word `"STOP"` is entered on a separate line.
- Next, a line is entered, where the names of the countries that are of interest to Ira are separated by a comma and a space.
- Information about some of these countries may not be included in the previously entered data.

#### 📤 OUTPUT FORMAT
- For each of the countries that Ira is interested in, the program does a check:
  - If the Gini index has been steadily increasing in the country (i.e., the value for 2018 was greater than for 2017; and the value for 2019 was greater than for 2018), then the program prints a phrase like `"<country name>: stable index growth"`.
  - If there was no stable increase in the Gini index in the country, then the program prints a phrase like: `"<country name>: fluctuations in the index values were observed"`.
  - If there is no data about such a country, the program prints a phrase like: `"<country name>: no data available"`.
  
- Next, a phrase of the form is displayed on a new line: `"A country with stable growth and the maximum index for 2019: <country name>"`.
  
- It is guaranteed that there is at least one country with stable index growth in the countries of interest to the Ira, and also that there is only one country with the highest index for 2019.


Шаги решения.

1) Считывание и обработка выходных данных.
2) Получение списка интересующих стран сравнение
3) Вывод результата

```python
# Шаг 1. Считывание и обработка входных данных

data = {} # пустой словарь для хранения пар страна:индексы по годам

while True: # бесконечный цикл для ввода данных
    info = input()
    if info == 'STOP': # если получили строку STOP, выходим из цикла
        break
    info = info.split(', ') # разделяем строку по запятой
    if len(info) == 1: # если в списке один элемент, значит это страна
        key = info[0] # сохраняем страну в переменную key
    else:
        data[key] = list(map(float, info)) # если элементов >1, значит это данные по годам. Преобразуем их в числа и сохраняем в словарь
```

```python
# Шаг 2. Получаем список нужных стран.

countries_to_check = input().split(', ') # сразу делим ввод и получаем список стран
```

```python
# Шаг 3. Считаем средние значения по странам.


maxi = 0 # переменная для хранения максимального значения индекса
for ctry in countries_to_check:
    if ctry in data: # проверяем, есть ли страна в словаре
        if data[ctry][1] > data[ctry][0] and data[ctry][2] > data[ctry][1]: # проверяем, растут ли индексы
            print(f'{ctry}: stable index growth')
            if data[ctry][-1] > maxi: # проверяем индекс 2019 года 
                maxi = data[ctry][-1]
                maxi_country = ctry # сохраняем страну с максимальным индексом
        else: # если индексы не растут, выводим сообщение
            print(f'{ctry}: fluctuations in the index values were observed')
    else: # если страны нет в словаре, выводим сообщение
        print(f'{ctry}: no data available')
        
print(f'A country with stable growth and the maximum index for 2019: {maxi_country}')
```

```python
# Полный код

data = {}

while True:
    info = input()
    if info == 'STOP':
        break
    info = info.split(', ')
    if len(info) == 1:
        key = info[0]
    else:
        data[key] = list(map(float, info))

countries_to_check = input().split(', ')

maxi = 0
for ctry in countries_to_check:
    if ctry in data:
        if data[ctry][1] > data[ctry][0] and data[ctry][2] > data[ctry][1]:
            print(f'{ctry}: stable index growth')
            if data[ctry][-1] > maxi:
                maxi = data[ctry][-1]
                maxi_country = ctry
        else:
            print(f'{ctry}: fluctuations in the index values were observed')
    else:
        print(f'{ctry}: no data available')
        
print(f'A country with stable growth and the maximum index for 2019: {maxi_country}')
```

  - It is guaranteed that information about this country is in the file.

#### 📤 OUTPUT FORMAT
- The names of the **industries that have generated the least income** in order from the lowest income to the highest.
  - Each name should be displayed on a **new line**.
  - It is guaranteed that there are **no industries with the same income** in the top three.

#### 📝 ADDITIONAL INFO
