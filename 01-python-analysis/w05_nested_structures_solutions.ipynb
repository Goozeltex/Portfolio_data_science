## 1. Registration for the hackathon

Many teams have registered for the student hackathon, and the organizers want to post a list of them on the website. Help them sort the list of team names alphabetically. Please note that letters in command names should not be case sensitive.

#### INPUT FORMAT

+ A string of command names separated by ";" (semicolon).

#### OUTPUT FORMAT

+ A string of command names separated by ";" (semicolon).

#### Step 1

We read a string of command names from the keyboard. We immediately make a list from it using the `.split()` method.

```python
 teams = input().split(';')
```

#### Step 2

Let's sort the resulting list with the key `key=str.lower` so as not to be case sensitive.

```python
teams = sorted(teams, key=str.lower)
```

#### Step 3

We output the sorted list in the required format using `*` (unpacking) and `sep=";"`.

```python
print(*teams, sep=';')
```

#### Final answer

```python
teams = input().split(';')
teams = sorted(teams, key=str.lower)
print(*teams, sep=';')
```

## 2. Vinyl records

Vasya collects music on vinyl records and wants to clean up his library by arranging the records by artist in alphabetical order, taking into account case.

#### INPUT FORMAT

+ Non-empty lines consisting of the title of the record and, separated by a semicolon and a space, the artist.
+ After all the lines with the names of the records there is a line “end”, which means the end of the input.
+ It is guaranteed that each performer occurs only once.

#### OUTPUT FORMAT

+ Lines with record titles, one title per line.
+ The order of display is determined by the alphabetical order of the artists (so David Bowie comes before Simon & Garfunkel and Queen).

#### Step 1

Create an empty dictionary.

```python
data = {}
```

#### Step 2

Let's implement the `while True` loop:
+ read from the keyboard into a new variable line with the name and artist.
+ we check the condition: if we counted the text `"end"`, then we break the loop using `break`.
+ if we have not left the loop, then we split the line from into two parts: title and artist using `.split()` (pay attention to the separator!).
+ write a pair into the dictionary, where the key is the artist, and the value is the name of the album.

```python
while True:
    info = input()
    if info == 'end':     # If the word "end" is entered, then
        break               # exit the loop
    info = info.split('; ')
    data[info[1]] = info[0] # Write down the key-value pair
```

#### Step 3

After completing the `while` loop, we implement a `for` loop for the sorted dictionary (dictionary keys): for each key we print its value.

```python
for key in sorted(data): # Don't forget about the sorted() function
    print(data[key])
```

#### Final answer

```python
data = {}

while True:
    info = input()
    if info == 'end':
        break
    info = info.split('; ')
    data[info[1]] = info[0]

for key in sorted(data):
    print(data[key])
```

## 3. Final exams

After final exams, the school prepares to print certificates. They can only be issued to students who achieve a passing grade. Prepare a list of students for whom certificates need to be printed.

#### INPUT FORMAT

+ A string listing students and their final exam scores, separated by semicolons and spaces. The student's name is separated from their exam score by a comma and a space.
+ It is guaranteed that information will be entered for at least one student.
+ On a new line, enter a positive integer - a passing grade.

#### OUTPUT FORMAT

+ List of student names, separated by commas and spaces, in order of appearance.

#### Step 1

We read a line with information about students from the keyboard. We split it by a semicolon with a space and save it into a variable - we get a list of strings. We also read the passing score into the variable and turn it into an integer.

```python
students = input().split('; ')
score = int(input())
```

#### Step 2

We create an empty list `res`, where we will add students who have scored a passing grade.

We run a `for` loop for the `students` list:
+ we split the student data using `split()` into last name and score - pay attention to the separator!
+ save the resulting objects into two variables (don’t forget to turn the score into an integer).
+ we check whether the student has a passing score or more (the `score` variable): if so, then we add his name to the `res` list.

```python
# Create an empty list
res = []

for st in students:
    st = st.split(', ')
    name = st[0]
    mark = int(st[1])
    if mark >= score:     # If the score is greater than or equal to the passing grade,
        res.append(name)  # add the student's name to the list
```

#### Step 3

We print the names from the `res` list, separated by a comma and a space. This can be done using the `*` operator and the `sep` parameter to the `print()` function.

```python
print(*res, sep=', ')
```

#### Final answer

```python
students = input().split('; ')
score = int(input())

res = []

for st in students:
    st = st.split(', ')
    name = st[0]
    mark = int(st[1])
    if mark >= score:
        res.append(name)
print(*res, sep=', ')
```

## 4. Products on the online store page

Mira chooses new furniture from an online store, but, unfortunately, she is limited in finances. Please help her choose by sorting all the products on the page from the most expensive to the cheapest.

#### INPUT FORMAT

+ Lines consisting of the name of the product and, separated by a semicolon and a space, its price in rubles. The price of each product is a positive integer.
+ It is guaranteed that information about at least one product will be entered.
+ After all lines with product names there is a line “end”, which means the end of the input.
+ It is guaranteed that all names are unique and all prices are unique.


#### OUTPUT FORMAT

+ Lines with product names, each name on a new line, in descending order of product price.

#### Step 1

Create an empty dictionary.

```python
data = {}
```

#### Step 2

Let's implement the `while True` loop:
+ read from the keyboard into a new variable line.
+ if the read line is “end”, we interrupt the loop.
+ if the cycle is not interrupted, then we divide the read line using `split()` into two lines - the name of the product and its price. Pay attention to the separator.
+ We add a pair to the dictionary: key - price (after all, we know that they are unique), turned into an integer, value - the name of the product.

```python
while True:
    info = input()
    if info == 'end':          # If the line "end" was inputted,
        break                    # interrupt the cycle
    info = info.split('; ')
    data[int(info[1])] = info[0] # Adding a key-value pair
```

#### Step 3

We start the `for` loop: we go through the `data` dictionary sorted by prices in descending order (don’t forget the `reverse=True` parameter!) - for each price key we print the corresponding value on a new line.

```python
for price in sorted(data, reverse=True):
    print(data[price])           # Printing dictionary values
```

#### Final answer

```python
data = {}

while True:
    info = input()
    if info == 'end':
        break
    info = info.split('; ')
    data[int(info[1])] = info[0]

for price in sorted(data, reverse=True):
    print(data[price])
```

## 5. Best courier

The head of the delivery service wants to find the best couriers among his employees - the courier with the highest average rating from customers.

Based on the results of each delivery, customers rated the couriers, and the manager has records of the performance of each courier on a scale from 1 to 5, where 1 is the worst delivery, 5 is the best.

Help the manager by reading delivery results records and finding the best courier in the company.

#### INPUT FORMAT

+ Lines consisting of the courier name and delivery estimate. The delivery estimate is separated from the courier name by a comma and a space and can only be an integer from 1 to 5.
+ It is guaranteed that at least one pair will be entered, consisting of the courier name and the rating.
+ After all lines with names there is a line “end”, which means the end of the input.

#### OUTPUT FORMAT

+ A line with the name of the best courier and his average score. It is guaranteed that there is only one best courier.
+ The name is separated from the GPA by a comma and a space.
+ There is a period at the end of the line.

#### Step 1

We create an empty dictionary - the names of couriers will be the keys, and the values ​​will be lists with their ratings.

```python
data = {}
```

#### Step 2

Run the `while True` loop:
+ Read information about the courier's name and his rating from the keyboard.
+ If the word “end” is given as input, then we interrupt the loop using `break`.
+ We save the courier’s name and his rating into separate variables (don’t forget to convert it to an integer).
+ We do a check - if the courier name is not in the `data` dictionary, then we create a pair in the dictionary: the courier name is a key, the value is a list with one element (its rating). If there is such a key, then we add a new rating to the value (remember that this is a list, so we use the `.append()` method) that corresponds to this key.

```python
while True:
    info = input()               # Enter name and score
    if info == 'end':          # If the input is the word "end",
        break                    # then we exit the cycle

    info = info.split(', ')
    name = info[0]
    score = int(info[1])

    if name not in data:         # If there is no such key,
        data[name] = [score]     # add it and its value as a list with one element
    else:                        # otherwise, we add to the existing list by key
        data[name].append(score) # to its value a new score
```

#### Step 3

Upon completion of the cycle, we access the key-value pairs in the `data` dictionary, calculate the average score for each value (list) and rewrite it in the dictionary.

```python
# Rewrite the key-value pairs (so that the values ​​are now
# represented the average score)
for k, v in data.items():
    data[k] = sum(v) / len(v)
```

#### Step 4

We find the maximum among the values ​​of the `data` dictionary. (we use the `.values()` method), save it to a variable.

```python
# Find the maximum value
maxi = max(data.values())
```

#### Step 5

We run a loop through the keys of the `data` dictionary and check: if the key value is equal to the maximum `maxi`, then we print the format string `"<name>, <average score>."`.

```python
for k, v in data.items():
    if v == maxi:
        print(f'{k}, {v}.')
```

#### Final answer

```python
data = {}

while True:
    info = input()
    if info == 'end':
        break

    info = info.split(', ')
    name = info[0]
    score = int(info[1])

    if name not in data:
        data[name] = [score]
    else:
        data[name].append(score)

for k, v in data.items():
    data[k] = sum(v) / len(v)

maxi = max(data.values())
for k, v in data.items():
    if v == maxi:
        print(f'{k}, {v}.')
```
