## 1. Date  
Write a program that takes a date in the format "DD.MM.YYYY" and outputs the same date in the format "DD month YYYY year".  

To substitute the textual names of the months, use the `months` list already provided in the code.  

#### INPUT FORMAT  
* A string containing a date in the format "DD.MM.YYYY", for example, "01.04.2017".  

#### OUTPUT FORMAT  
* The entered date in the format "DD month YYYY year", for example, “1 April 2017 year”.

#### Step 1

List with months `months` we already have. And as input we will have string with date. 

```python
months = ['января', 'февраля', 'марта', 
          'апреля', 'мая', 'июня', 
          'июля', 'августа', 'сентября', 
          'октября', 'ноября', 'декабря']

date_str = input()
```

#### Step 2

We split our date string by dots using the .split('.') method, which results in a list containing the day, month, and year.

We convert the month number into an integer and use it as an index to look up the month in the months list. Don't forget to subtract one before doing this, since indexing starts at zero.

We then replace the month number in the list with its name.

```python

# Split the date string by dots, resulting in a list ['day', 'month', 'year']
date_list = date_str.split('.')  

# Take the month number from this list and convert it to an integer
month_num = int(date_list[1])  

# Get the month name from the months list using the index, which is the month number - 1
month_str = months[month_num - 1]  

# Replace the element at index 1 in the original list, 
# changing the month number to its word representation
date_list[1] = month_str  
```

#### Step 3

Создаем новую строку с датой в нужном формате с помощью f-строки. Выводим эту строку на печать. 

```python
# Creating a new variable that will have date in the desirable format
date_str_new = f'{date_list[0]} {date_list[1]} {date_list[2]} года'

# Printing new string
print(date_str_new) 
```

#### Final solution

```python
months = ['января', 'февраля', 'марта', 
          'апреля', 'мая', 'июня', 
          'июля', 'августа', 'сентября', 
          'октября', 'ноября', 'декабря']

date_str = input()

date_list = date_str.split('.')

month_num = int(date_list[1])

month_str = months[month_num - 1]

date_list[1] = month_str

date_str_new = date_list[0] + ' ' + date_list[1] + ' ' + date_list[2] + ' года'

print(date_str_new)
```

## 2. Dictation  
The teacher is composing a dictation to test how well the students have learned a new spelling rule. Write a program that will check how many words in the dictation match the given rule.

#### INPUT FORMAT  
* The first line contains letter combinations that should appear in the dictation words, separated by spaces.  
* The second line contains the dictation text: all sentences are written in lowercase letters, without capitals.  
* It is guaranteed that each word does not contain two or more different target letter combinations.

#### OUTPUT FORMAT  
* A single number — how many words in the dictation text contain the required letter combinations.

#### Step 1

At the first step, we take the letter combinations and the dictation text as input. We use the `.split()` method to immediately convert the input strings into lists. <br>  
We also create a counter variable to count the matching words in the text.

```python
letters = input().split()
dictation_words = input().split()
words_ok = 0
```

#### Step 2

Next, we will check for the presence of each combination in the dictation words. The first loop will output the dictation words. We'll print them for clarity (we won't print them in the final solution).

```python
for word in dictation_words:  # For each word in dictation —
    print(word)               # Print this word
```

#### Step 3

The nested loop will iterate through the combinations. We'll print them for clarity.

```python
for letter in letters: 
    print(letter)
```

#### Step 4

Finally, we combine the nested loops. For each word in the dictation text, we check every letter combination, and if it is present in the word, we add +1 to our counter variable.

```python
for word in dictation_words:
    for letter in letters:      
        if letter in word:
            words_ok += 1
```

#### Final solution

```python
letters = input().split()
dictation_words = input().split()
words_ok = 0

for word in dictation_words:
    for letter in letters:      
        if letter in word:
            words_ok += 1
print(words_ok)
```

## 3. Household Chores — 1  
Zhenya and Sasha are dividing household chores from a shared list, alternating tasks: Zhenya takes the first, third, and so on, while Sasha takes the second, fourth, and so on. Write a program that takes the shared list of chores and outputs the task lists for Sasha and Zhenya.

#### INPUT FORMAT  
* A non-empty shared list of chores, separated by commas and spaces.

#### OUTPUT FORMAT  
* Two lists separated by commas and spaces: first Sasha's list of chores, then Zhenya's list of chores.  
* Each list is printed on a new line.

#### Step 1

First, we take the input string containing all the chores. <br> We split it into individual tasks using the `.split(', ')` method. <br> We then create empty lists for Sasha and Zhenya, where we will assign each of their tasks.

```python
task_str = input()

task_list = task_str.split(', ')

s_tasks = []
z_tasks = []
```

#### Step 2

We record Sasha's tasks into his list. To do this, we use a loop and the `range()` function, which will iterate through the indices of the task list.

**The first** argument is the index of the element where we start iterating — in our case, it's `[1]`, since Sasha gets the tasks starting from the second one (and indexing in Python starts at zero). **The second** argument is the end of the sequence — the length of the task list. **The third** argument is the step, i.e., which element we should pick after the previous one — since tasks are distributed alternately, the step will be 2.

Each second element, starting from the second one, is added to Sasha's list using the `.append()` method, which appends a new element with index `i` to the current list.



```python
for i in range(1, len(task_list), 2):
    s_tasks.append(task_list[i])

# Printing out Sasha's list to check the output
print(s_tasks)  
```

#### Step 3

Next, we need to add Zhenya's tasks to his list. We use the same `range()` construct. **The first** argument is `0`, meaning the tasks starting from the first one in the list will be Zhenya's. **The second** argument is also the end of the sequence. **The third** is the same step of 2, meaning every other element.

Just like for Sasha, at each step of the loop, we append the element with this index to Zhenya's list.

```python
for i in range(0,len(task_list),2):
    z_tasks.append(task_list[i])

# Checking the output
print(z_tasks) 
```

#### Step 4

At the end, we print both lists in sequence. For each list, we use `sep=', '`, since the tasks need to be displayed separated by a comma followed by a space.

```python
print(*s_tasks, sep=', ')
print(*z_tasks, sep=', ')
```

#### Final solution

```python
task_str = input()

task_list = task_str.split(', ')

s_tasks = []
z_tasks = []

for i in range(1, len(task_list), 2):
    s_tasks.append(task_list[i])
    
for i in range(0, len(task_list), 2):
    z_tasks.append(task_list[i])
    
print(*s_tasks, sep=', ')

print(*z_tasks, sep=', ')
```

## 4. Library  
Zhenya has a list of textbooks that need to be borrowed from the library, along with their corresponding ISBN numbers (a book and its corresponding number are in two lists at the same position). To find a book in the library, the ISBN must be provided. Some books Zhenya already owns and do not need to be borrowed.

Write a program that takes a list of book titles, then a list of ISBN numbers, then the titles of books Zhenya already owns, and outputs the ISBN numbers of the books Zhenya doesn't have yet.

#### INPUT FORMAT  
* Exact book titles from the textbook list, respecting case, separated by commas and spaces.  
* ISBN numbers of the textbooks, respecting case, separated by commas and spaces.  
* Exact book titles that Zhenya already owns, respecting case, separated by commas and spaces. Zhenya may not own any textbooks.

#### OUTPUT FORMAT  
* ISBN numbers of the books Zhenya doesn't own, each number on a new line.

#### Step 1

At the first step, we take the input of the list of all book titles, the list of all ISBN numbers, as well as the books Zhenya already owns. For each line, we apply the `.split(', ')` method, which splits the string into elements by a comma and a space, resulting in lists.

```python
books_all_title = input().split(', ')

books_all_isbn = input().split(', ')

books_own = input().split(', ')
```

#### Step 2

We need to iterate through the book titles available in the library and check for each one whether Zhenya already owns it.

We will use loops. We iterate over the indices using a `for` loop and the `range()` function, where the argument is the length of the list of all books. On each iteration, we will store the index of the current book in the variable `idx`.

The construct `books_all_title[idx]` returns the title of the book, which we can then check against the list of owned books. We'll print the result for clarity.

```python
for idx in range(len(books_all_title)):
    print(books_all_title[idx])
```

#### Step 3

At each step of the loop, we use an `if` condition to check that the book title we get by the index `[idx]` from the list of all books is NOT in the list of books that Zhenya already owns. If the condition is true, we print the ISBN of that book, retrieving it by the same index `[idx]` — since the ISBN and book titles are arranged in the same order in both lists.

```python
for idx in range(len(books_all_title)):
    if books_all_title[idx] not in books_own:
        print(books_all_isbn[idx])
```

#### Final solution

```python
books_all_title = input().split(', ')

books_all_isbn = input().split(', ')

books_own = input().split(', ')


for idx in range(len(books_all_title)):
    if books_all_title[idx] not in books_own:
        print(books_all_isbn[idx])
```

## 5. Email Addresses  
Taisia is updating the email addresses of her classmates before sending out a mailing. Classmates send her their unused email address and the one they want to replace it with, separated by a space. Taisia finds the address in her list and replaces it with the new one.

#### INPUT FORMAT  
* The email addresses of Taisia's classmates, separated by commas.  
* The number of classmates whose addresses need to be replaced — a non-negative integer. If there are no such classmates (0 is entered), the next step is skipped.  
* For each classmate, a pair "<old address> <new address>" is read, separated by a space.

#### OUTPUT FORMAT  
* The final list of addresses, with the addresses separated by semicolons.

#### Step 1

```python

# Input all emails and split them by commas to get a list
emails = input().split(',')

# Input the number of addresses that need to be changed
n = int(input())
```

We receive all the email addresses as input, use the `.split(',')` method to split the string by commas and convert it into a list. <br> We then read `n` — the number of addresses that need to be changed. <br>

#### Step 2

We start a loop to change the addresses for the required number of students. We take the current email address that needs to be changed and the one it should be replaced with.

We will read the pairs in a loop of the type `for i in range(n)` — since the number `n` specifies the number of address pairs we need to process.

```python
for student in range(n): 
    to_change = input().split()

# Checking out the output
print(to_change) 
```

If, for example, you input `n` greater than one, the last read pair will be printed.

#### Step 3

Next, we need to find the address that needs to be replaced in Taisia's original list. We'll implement a nested loop.

The nested loop works like this: we iterate over all current addresses in the list and check if the address matches the one that needs to be changed (the first in the `to_change` list). If it matches, we replace the old address with the new one (the second in the `to_change` list). Then, we end the loop with the `break` function.

```python
for i in range(len(emails)):      # Iterate over the indices of all addresses
    if emails[i] == to_change[0]: ## if the address in the list matches the one that needs to be changed,
        emails[i] = to_change[1]  ## replace it with the new one, updating the list
        print(emails)             ## print the list to see the result after each address change
        break
```

#### Step 4

Finally, we need to print all the addresses separated by semicolons, using the `sep` parameter for this.

```python
print(*emails, sep=';')
```

#### Final solution

```python
emails = input().split(',')
n = int(input())

for student in range(n):
    to_change = input().split()
    for i in range(len(emails)):
        if emails[i] == to_change[0]:
            emails[i] = to_change[1]
            break
            
print(';'.join(emails))
```
