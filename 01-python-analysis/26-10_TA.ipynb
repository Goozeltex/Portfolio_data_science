# Q 15

Anastasia analyzes information about the acting fees of the actors of the series. Write a program that will help her organize the information and identify the actor who received the highest fee.

#### INPUT FORMAT
- A text file stars_budget.txt, which records data on the series and the royalties of their cast. For example, "Once Upon a Time;Ginnifer Goodwin-550;Jennifer Morrison-293". In this example, "Once Upon a Time" is the name of the series; "Ginnifer Goodwin-550" is the name of the actor who received a fee of 550; "Jennifer Morrison-293" is the name of the actor who received a fee of 293.
- It is guaranteed that the fee amount is always an integer.
- It is guaranteed that the names of the actors in the file are not repeated.
- The separator in the file is a semicolon. The title of the series or the name of the actor can contain any characters except semicolons.
- It is guaranteed that information about only one series is recorded on one line.
- It is guaranteed that there is at least one data line in the file.

#### OUTPUT FORMAT
- The program should determine the name of the series whose actor received the highest acting fee, the name of this actor and the amount of the fee. It is also necessary to determine the fee of all the actors of this series.
The result of the program should be a text file result.txt with lines of the following format:
- "The series where the actor received the highest fee: <series name>". It is guaranteed that there is only one actor with such a fee in the data.
- Next, a line with the actor's name and the amount of the fee is displayed in the format "Actor's name: <actor's name>, fee amount: <fee amount>". 
- Next, the string "===" is output once.
- Then the program prints the phrase "The names of the actors of the series and their fee:" and determines how much each actor of this series earned. A phrase in the format "<actor's name>, <fee amount>" is printed from a new line for each actor
- The names of the actors in the series should be in the same order as in the original file. Each entry must be printed from a new line.

#### ADDITIONAL INFO
- The open test uses files stars_budget.txt and result.txt. You can download them and test your solution locally.
- In the example, the test number is displayed in the first column (0 is an open test), you do not need to read this number or use it in your solution.

### Шаги решения

* Считать и разобрать данные из файла stars_budget.txt，организовать данные по сериалам и актёрам.
* Определить сериал с актёром, получившим наибольший гонорар, а также имя этого актёра и сумму гонорара.
* Вывести результаты в указанном формате в файл result.txt.



```python
# Шаг 1. Чтение и разбор данных

DATA_S = {} # создаем словарь для данных сериалов
with open('stars_budget.txt', 'r', encoding='utf-8') as file:
    
    for line in file: # перебираем по строчкам
        data_a = {} # создаем словарь для данных актеров
        line = line.strip().split(';') # разбиваем строку на список. Первый эелемент - название сериала. Второй - список актеров с гонорарами.
        
        for actor in line[1:]:
            actor, money = actor.split('-')
            data_a[actor] = int(money) # сохраняем в словарь парами
            
        data_a['Max'] = max(data_a.values()) # добавляем пару ключ:значение с максимальным гонораром
        DATA_S[line[0]] = data_a # Добавляем в словарь пару сериал:словарь актеров и гонораров

print(data_a)
print(DATA_S.items())
```

```python
# Определяем сериал, где актер получил наибольший гонорар

maxi_star_k = list(DATA_S.keys())[0] # берем самый первый сериал
maxi_star_b = DATA_S[maxi_star_k]['Max'] # берем максимальное значение этого сериала

for k, v in DATA_S.items(): # Перебираем по парам сериал:словарь актеров и гонораров
    if v['Max'] > maxi_star_b: # Если максимальное значение больше, чем текущее максимальное значение
        maxi_star_k = k # Перезаписываем сериал
        maxi_star_b = v['Max'] # Перезаписываем максимальное значение

# Таким образом мы последовательно сравним разные сериалы и найдем сериал с максимальным гонораром

# Мы получили две переменные, с сериалом где был максимальный гонорар и сам размер гонорара
```

```python
# Определяем имя актера, получившего максимальный гонорар
print(DATA_S[maxi_star_k].items()) # посмотрим, по чему перебираем

for k, v in DATA_S[maxi_star_k].items(): # Перебираем по парам актер:гонорар
    if v == maxi_star_b and k != 'Max': 
        actor = k # сохраняем имя актера
```

```python
# Выводим результат

with open('actors.txt', 'w', encoding='utf-8') as file:
    print(f'The series where the actor received the highest fee: {maxi_star_k}', file=file)
    print(f"Actor's name: {actor}, fee amount: {maxi_star_b}", file=file)
    print('===', file=file)
    print('The names of the actors of the series and their fee:', file=file)
    for k, v in DATA_S[maxi_star_k].items():
        if k != 'Max':
            print(f'{k}, {v}', file=file)
```

```python
# Весь код


DATA_S = {}
with open('stars_budget.txt', 'r', encoding='utf-8') as file:
    
    for line in file:
        data_a = {}
        line = line.strip().split(';')
        
        for actor in line[1:]:
            actor, money = actor.split('-')
            
            data_a[actor] = int(money)
            
        data_a['Max'] = max(data_a.values())
        DATA_S[line[0]] = data_a
        
maxi_star_k = list(DATA_S.keys())[0]
maxi_star_b = DATA_S[maxi_star_k]['Max']

for k, v in DATA_S.items():
    if v['Max'] > maxi_star_b:
        maxi_star_k = k
        maxi_star_b = v['Max']

for k, v in DATA_S[maxi_star_k].items():
    if v == maxi_star_b and k != 'Max':
        actor = k
        
with open('result.txt', 'w', encoding='utf-8') as file:
    print(f'The series where the actor received the highest fee: {maxi_star_k}', file=file)
    print(f"Actor's name: {actor}, fee amount: {maxi_star_b}", file=file)
    print('===', file=file)
    print('The names of the actors of the series and their fee:', file=file)
    for k, v in DATA_S[maxi_star_k].items():
        if k != 'Max':
            print(f'{k}, {v}', file=file)
```

# Q 14

### Задание

Kirill has been analyzing public opinion about his favorite TV channels for several years. He wants to understand which channel is the most popular in people's preferences. Write a program that will help him with this task.

#### 📥 INPUT FORMAT
- An integer is entered — the number of years for which the data will be entered.
- Next, entries are entered with information about preferences for TV channels for each year in the form `"<year>: <names of TV channels separated by commas and spaces>"`. Each entry is entered from a new line.
- The number of TV channels is always the same, as are their names. For example, if in the first year there were three TV channels — TNT, Russia 1 and CTC, then in all other years there will be the same TV channels, but perhaps in a different sequence.
- Each line contains information for only one year.
- It is guaranteed that the names of TV channels are not repeated in one line.

Kirill uses the following method to analyze preferences for TV channels:
- Each TV channel is assigned a **rank** depending on its position in preferences.
- For example, if the TV channels are arranged as follows: `"TNT, Russia-1, CTC"`, then the rank of TNT is 3, Russia-1 is 2, and CTC is 1.
- The ranks of the TV channel for each year are summed up. As a result, the TV channel with the highest sum of the ranks is considered the most popular.

#### 📤 OUTPUT FORMAT
- The name of the TV channel that is the most popular, according to the applied method.
- It is guaranteed that there is only one such TV channel.


### Шаги решения

1) Считываем и разбиваем данные
2) Анализируем данные
3) Выводим информацию


```python
# Шаг 1. Ввод данных

N = int(input()) # принимаем количество лет, поэтому сразу преобразуем в int

data = {} # создаем словарь для данных


while N != 0: # пока не закончатся года
    year, channels = input().split(': ') # делим ввод на два списка
    
    data[year] = channels.split(', ') # список каналов разбиваем на отдельные элементы и сохраняем в словарь
    
    N -= 1 # уменьшаем счетчика годов
```

```python
# Шаг 2. Анализируем и считаем рейтинг.

result = {} # создаем словарь для рейтинга

for k, v in data.items(): # k - год, список каналов за этот год
    for i in range(len(v)): # перебираем каналы по индексу
        if v[i] not in result: # если канала нет в словаре
            result[v[i]] = i # добавляем в словарь пару канал:индекс
        else:
            result[v[i]] += i # если канал есть, увеличиваем его ключ на индекс

# Важно! В нашем случае у первого канала по списку рейтинг будет +0
# Поэтму канал с минимальным рейтингом будет лучшим
```

```python
# Определяем канал с наименьшей суммой рейтингов

mini = min(result.values()) # находим минимальное значение и сохраняем в переменную

for k, v in result.items(): # перебираем по парам канал:рейтинг
    if v == mini: # если рейтинг равен минимальному
        print(k) # выводим канал
```

```python
# Полный код

N = int(input())

data = {}

while N != 0:
    year, channels = input().split(': ') # year: channel
    
    data[year] = channels.split(', ')
    
    N -= 1
    
result = {}

for k, v in data.items():
    for i in range(len(v)):
        if v[i] not in result:
            result[v[i]] = i
        else:
            result[v[i]] += i

mini = min(result.values())

for k, v in result.items():
    if v == mini:
        print(k)
```

## Задание

Lyuba studies information about films and wants to determine the top three films by the number of views among films of the **criminal genre** with a certain duration. Help Lyuba make a list of top films.

### 📥 INPUT FORMAT
- A text file `films.txt`, which contains information about the films in the format `"<movie name>;<duration>;<rating>;<genres of the film separated by commas and spaces>;<number of views>"`.
  - The **semicolon** is used as a separator in the file.
  - Each line in the file contains information about only **one movie**.
  - It is guaranteed that there are **no empty lines** in the file.
- The **duration** of the film (an integer) is also introduced, specifying that Lyuba is interested in films of the **criminal genre** (the `"Crime"` genre in the data) with a duration **strictly longer** than specified.

#### 📤 OUTPUT FORMAT
- The names of films of the criminal genre (`"Crime"` in the data) with a duration longer than specified, sorted in the **top three** in terms of the **number of views** among films that meet these conditions.
- Pay attention to the format of recording the number of views. For example, the string `"3,788"` means that the movie has been watched **3,788 times**.
- It is guaranteed that there are **no films in the top three** with the same number of views.

#### 📝 ADDITIONAL INFO
- The open test uses a file `films.txt`. You can download it and test your solution locally.
- In the example, the test number is displayed in the first column (0 is an open test); you **do not need** to read this number or use it in your solution.


```python
# Считывание и обработка данных

mins = int(input()) # минимальная продолжительность фильма в минутах

data = {} # создаем словарь для данных

with open('films(26_13).txt', 'r', encoding='utf-8') as file: 
    for line in file: # перебираем по строчкам
        line = line.strip().split(';') # разбиваем строку на список по разделителю ;
        f_mins = int(line[1].split()[0]) # берем второй элемент списка, разбиваем по пробелу и берем первый элемент
        if f_mins > mins and 'Crime' in line[3]: # 
            name = line[0] # берем первый элемент из строчки (название фильма)
            rating = int(line[-1].replace(',', '')) # берем число просмотров заменяя запятую на пустоту
            
            data[name] = rating # сохраняем в словарь пару название фильма:рейтинг
```

```python
# Сортировка и вывод

res = [] # создаем список для результата
for k, v in data.items(): # перебираем по парам название фильма:рейтинг
    res.append((v, k)) # добавляем в список пару рейтинг:название фильма (чтобы сортировать по рейтингу)
    
for film in sorted(res, reverse=True)[:3]: # сортируем список по убыванию и перебираем по первым трем эелементам
    print(film[-1]) # выводим название фильма
```

```python
# Полный код

mins = int(input())

data = {}

with open('films(26_13).txt', 'r', encoding='utf-8') as file:
    for line in file:
        line = line.strip().split(';')
        f_mins = int(line[1].split()[0])
        if f_mins > mins and 'Crime' in line[3]:
            name = line[0]
            rating = int(line[-1].replace(',', ''))
            
            data[name] = rating
            
res = []
for k, v in data.items():
    res.append((v, k))
    
for film in sorted(res, reverse=True)[:3]:
    print(film[-1])
```
