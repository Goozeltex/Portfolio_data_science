## 1. Coursework
A line is given that describes the committee's decision regarding the course work. Print "YES" if the solution begins with the substring "pass", or "NO" otherwise. Please note that the solution can be written in any case (for example, capital letters, small letters, or capital letters).

#### INPUT FORMAT
* Non-empty string

#### OUTPUT FORMAT
* String "YES" or "NO"

#### Step 1

First we read the string using the `input()` function. Next, using the `.lower()` method, we convert the entire string to lowercase.

 We see that the line has indeed become all lowercase.


```python
decision = input()
decision = decision.lower() # We get the commission’s decision as input and convert it to lower case
print(decision, '\n')
```

It is also important to know the `.islower()` method, which can be used to check if the entire string is lower case. It returns `True` if the string is lowercase, and `False` if it is not. This method is not needed for solving this problem, but often this operation can be useful.

```python
print('hello'.islower())
print('Goodbye'.islower())
```

#### Step 2

We create an `if` loop, check the condition that the line begins with the phrase “pass”, use the `.startswith('pass')` method for this. Depending on this, we display “YES” or “NO”.

```python
if decision.startswith('pass'): # We check the condition that the line begins with “pass”
    print('YES')
else:
    print('NO')
```

#### Final answer

```python
decision = input().lower()
if decision.startswith('pass'):
    print('YES')
else:
    print('NO')
```

## 2. Olympiad
N people took part in the Olympiad. Their results are written in the format 'M points', and there can be any number of spaces at the beginning and end of the line. Calculate the average score of the participants.

#### INPUT FORMAT
* The program receives as input a positive integer number of Olympiad participants N.
* Next comes N lines, each line contains the result of the participant in the format described above.

#### OUTPUT FORMAT
* Real number is the average score of all participants.

#### Step 1

First we take `N` as input - the number of students. We create a variable where we will add the sum of all points.

```python
N = int(input())
total = 0
```

#### Step 2

Students' scores will be sent to us in the form of strings in the format `' 9 points '`. After reading, we immediately process this line: we use the `.strip()` method, which trims all spaces and whitespace characters at the beginning and end of the line.

Next, we use the already familiar `.split()` method, which divides the string into `"X"` elements - a number and `"points"`. We take the first element from this list with index `[0]` - the number of points. Then we turn it into an integer and add it to the total points.

For clarity, we will demonstrate all the applied methods on line `a`.

```python
example = '   9 points   '         # String of the required format
print(example.strip())             # Print a string without spaces
print(example.strip().split())     # Turn a string without spaces into a list
print(example.strip().split()[0])  # We get the number of points
```

Now we count the data - create a `for` loop using `range(N)`, which will run `N` times. For each student, we take as input a string in the format `"X points"`, with any number of spaces at the ends, and extract the number of points from it.

```python
for i in range(N):
    points = input().strip().split()[0]  # Cut by spaces, divide into elements “N-number” and “points”
    print(int(points))                   # Print the number
```

Now we will not print the points, but add them to our `total` variable.

```python
for i in range(N):
    points = input().strip().split()[0]
    total += int(points)
```

#### Step 3

Divide the total by the number of students to get the average score.

```python
print(total / N)
```

#### Final answer

```python
N = int(input())
total = 0
for i in range(N):
    points = input().strip().split()[0]
    total += int(points)
print(total / N)
```

## 3. Tests
Petya solves tests from test materials and writes down answers in the form of a four-digit question number and the answer in the form of a single capital letter (for example, 4578A). Answers are separated by commas and spaces. At the end of the book there are answers to questions given in the form of a sequence of capital letters. Help Petya compare whether he solved the test correctly.

#### INPUT FORMAT
* List of answers separated by commas and spaces.
* String is a sequence of capital letters of the Latin alphabet with correct answers.

#### OUTPUT FORMAT
* Two lines:
    * "Your answer: " and a sequence of capital letters of the Latin alphabet.
    * If the test is solved correctly, “Well done! Everything is correct” is displayed, and if it is incorrect - “The test is solved incorrectly, the correct answer is: “ and the sequence of capital letters of the Latin alphabet of the correct answer.

#### Step 1

We receive a string with numbers and answers as input, split it by comma and space using the `.split(', ')` method, and get a list. We create a new empty list where we will write down Petya’s answers.

```python
answers_petya = input().split(', ')
keys = input()

answers_letters = []
```

#### Step 2

For each answer in the `for` loop, we take the last element - the letter of the answer - and add it to a new list. The `[-1]` index is perfect for this task.

```python
for i in answers_petya:
    answers_letters.append(i[-1]) # For each Petya's answer we take the last element - the letter of the answer

print(answers_letters)            # This is the list we got
```

#### Step 3

Next, using the `.join()` method, we put all the letters of Petya’s answers into one line.  Let's print it to check what happened.

```python
answers_string = ''.join(answers_letters) # We combine all the letters of the answers into one line
print('Your answer:', answers_string)
```

#### Step 4

Using the `if` condition, we check whether the answers are correct. Depending on the correctness of the answers, we print one or another line.

```python
if answers_string == keys: # Compare with correct answers
    print('Well done! Everything is correct')
else:
    print('The test is solved incorrectly, the correct answer is:', keys)
```

#### Final answer

```python
answers_petya = input().split(', ')
keys = input()

answers_letters = []

for i in answers_petya:
    answers_letters.append(i[-1])

answers_string = ''.join(answers_letters)
print('Your answer:', answers_string)

if answers_string == keys:
    print('Well done! Everything is correct')
else:
    print('The test is solved incorrectly, the correct answer is:', keys)
```

## 4. Corporate emails
The Pamparam company registers corporate emails for employees and wants to leave them with the logins that they are already accustomed to using. The only condition is that mail should not begin with numbers or with a dot, dash, or underscore, so if such characters are found, they must be discarded. Help the company create a new mail list on the pamparam.ru domain

#### INPUT FORMAT
* List of mailing addresses of company employees separated by semicolons and spaces.
* It is guaranteed that each address contains at least one letter.

#### OUTPUT FORMAT
* List of new mailing addresses for company employees - domains (what comes after @) should be replaced with the pamparam.ru domain, and on the left side the characters that should not begin with the mailing address should be discarded.
* List the list separated by semicolons and spaces.

#### Step 1

First, we take all addresses as input, divide them by `';'` and a space using the `.split('; ')` method, and get a list with addresses. We create a new list where we will add the corrected addresses.

```python
old_emails = input().split('; ') # We get strings with emails as input and separate them by '; '
new_emails = []
```

#### Step 2

First, let's experiment with our transformations on a separate address. First, we cut off the characters and numbers at the beginning of the address using the `.lstrip()` method. Then we split by the `'@'` sign, using `.split('@')`, take the login (before the "dog"), and add a new domain to it - `'@pamparam.ru'`.

```python
email = '123L@hse.ru' # Old address

# We cut off prohibited characters and numbers at the beginning of the email address
print(email.lstrip('_-.0123456789'))

# Split adress by @
print(email.lstrip('_-.0123456789').split('@'))

# Retrieving the login
print(email.lstrip('_-.0123456789').split('@')[0])

# Adding a new domain
print(email.lstrip('_-.0123456789').split('@')[0] + '@pamparam.ru')
```

Now in a loop we will perform these operations for all addresses. We write the new corrected address to the created list using `.append()`.

```python
for email in old_emails:
    email = email.lstrip('_-.0123456789')        # Update the email variable with an address without prohibited characters
    email = email.split('@')[0] + '@pamparam.ru' # Let's update the variable with the address with the new domain
    print(email)                                 # Let's print the corrected address for clarity
    new_emails.append(email)                     # Add this address to the list

```

Let's see what the list of new addresses looks like.

```python
print(new_emails)
```

#### Step 3

Finally, using the `.join()` method, we print the corrected addresses.

```python
print('; '.join(new_emails)) # We print the list with '; '
```

#### Final answer

```python
old_emails = input().split('; ')
new_emails = []

for email in old_emails:
    email = email.lstrip('_-.0123456789')
    email = email.split('@')[0] + '@pamparam.ru'
    new_emails.append(email)

print('; '.join(new_emails))
```

## 5. Sports season
During the sports season, 5 competitions are held, based on the results of which points are awarded:
* If the athlete wins, then 5 rating points are awarded.
* If he/she didn’t win, but were in the top three - 3 points.
* For participation without a prize place, 1 point is awarded.

Write a program that, based on the standings and the name of the athlete, calculates the rating (sum of points) of this athlete.
#### INPUT FORMAT
* Tournament table - five lines, in each of which, separated by commas, the names of the athletes in the order of what places they took in the corresponding competition.
* After this, another line is entered - the name of the athlete

#### OUTPUT FORMAT
* Integer - rating of the entered athlete

#### Step 1

We create a list where we will record all the results of the competition. We also create a counter variable, where there will be points for our requested athlete.

Next, using the `for i in range(5)` loop, we take as input a string with the names of the athletes in the order of their results 5 times and add them to the list.  

Then we read the last name of the athlete for whom we need to display the rating.

```python
table = []

for i in range(5):         # For each of the 5 competitions
    athletes = input()     # We accept a line with athletes according to their places
    table.append(athletes) # Adding a string to the list

# For clarity, we will print the list in the last step
print(table)

# We receive as input the name of the athlete for whom we need to display the rating
athlete = input()
```

#### Step 2

Next you will have to work with indexes. Let's remember how the `.index()` method works, which, based on a value from a list, displays its index.

```python
example = ['Pupkin', 'Dubrovkin']
print(example.index('Pupkin'))
print(example.index('Dubrovkin'))
```

Accordingly, to find out what place the athlete took, we will find his index in the list. The main thing to remember is that Python counts from 0 and the first place will be "zero".

Now we need to find out whether our athlete participated in the competition and what place he took. Let's create a counter variable `points` to count points (initially equal to 0).

Then let's loop through our list of strings:
* Using `split()` we turn the string into a list of surnames (split by comma with a space) and save the list into a variable.
* We check the condition to see if our athlete is on this list. If so, then we check the nested conditions:
    * if the index of this surname in the list is zero (first place), then add 5 to the counter variable,
    * if not, then we check whether the index is equal to 1 or 2 (second or third place) - in this case we add 3,
    * in all other cases we add 1 (after all, we have already checked that the athlete is on the list, which means he participated in the competition).

```python
points = 0

for event in table:
    event = event.split(', ')                  # We get a list of surnames
    print(event)                               # Let's print it for clarity
    if athlete in event:                       # We check whether athlete participated
        if event.index(athlete) == 0:          ## If he took 1st place, then...
            points += 5
        elif event.index(athlete) in (1, 2):   ## If he took 2nd or 3rd place, then...
            points += 3
        else:                                  ## If he/she isn't among the top three, but participated, then...
            points += 1

print(points)
```

#### Final answer

```python
table = []

for i in range(5):
    athletes = input()
    table.append(athletes)

athlete = input()

points = 0

for event in table:
    event = event.split(', ')
    if athlete in event:
        if event.index(athlete) == 0:
            points += 5
        elif event.index(athlete) in (1, 2):
            points += 3
        else:
            points += 1

print(points)
```
