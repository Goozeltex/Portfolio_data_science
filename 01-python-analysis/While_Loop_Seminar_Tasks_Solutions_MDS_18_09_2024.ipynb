## 1. Lucky numbers
Oleg considers his lucky numbers to be all those divisible by 4 and 6 at the same time (for example, 12). Oleg wants to write a program that takes a number as input and displays all of Oleg's lucky numbers in the range from 1 to this number inclusive. If there are no such numbers, then the program does not output anything.

#### INPUT FORMAT
* Positive integer N.

#### OUTPUT FORMAT
* All numbers in the range from 1 to N inclusive, which are lucky for Oleg, each on a new line.

#### Step 1

We enter the number `number` from the keyboard, create a variable `i`, which is a counter number, which we will check at each step.

Next, we set up a `while` loop that runs as long as `i <= number`.
To check how the loop works, let's increase `i` by 1 so that at some point the loop will break (when `i` becomes greater than `number`). We will also print `i`.

```python
number = int(input()) # Reading the number
i = 1                 # Making a counter variable

while i <= number:     # While the value of the counter variable
                       # did not exceed the specified number -
    print(i)           # print the number i
    i += 1             # increase i by 1
```

#### Step 2

Now let's return to the problem statement - we need to print only the “lucky” values ​​of `i`. At each step of the loop, we check through the `if` condition whether `i` is divisible by 4 and 6 at the same time. If it is divided, then we print it. At the end of the iteration we increase `i` by 1.

```python
number = int(input()) # Reading the number
i = 1                 # Making a counter variable

while i <= number:    # While the value of the counter variable
                      # did not exceed the specified number -
    if i % 4 == 0 and i % 6 == 0:  # if divisible by 4 and 6 without remainder, then
        print(i)                   # print the counter variable

    i += 1                         # Increase the value of the counter variable by 1
```

#### Final answer

```python
number = int(input())
i = 1

while i <= number:
    if i % 4 == 0 and i % 6 == 0:
        print(i)
    i += 1
```

## 2. Increased scholarship - 1
The increased scholarship for excellent studies is equal to 20,000. Vasya decided to calculate whether he would have enough money to live if he received a scholarship. He calculated and recorded his expenses for the last 12 months. Help Vasya find the number of months when he clearly would not have had enough scholarship.

#### INPUT FORMAT
* 12 non-negative integers, each on a new line.

#### OUTPUT FORMAT
* A number indicating the number of months when Vasya spent more than 20,000 rubles per month.

#### Step 1

We create a counter variable to go through the loop and a month counter that will satisfy the condition. <br>

```python
month = 0 # Number of months to calculate expenses
count = 0 # Number of months when Vasya spent more than 20,000 rubles
```

#### Step 2

We make a `while` loop that runs as long as the number of months checked is less than 12 (that is, the loop is executed 12 times, which corresponds to 12 months). For now, we'll just run a loop that runs 12 times.



```python
month = 0
count = 0

while month < 12: # 12 months, going through every of them one by one
    month += 1

print(month)      # We print to check on which month we left the cycle
```

#### Step 3

Now we add a condition check. Inside the loop we will read the amount of monthly expenses.

If the amount of expenses is more than 20,000, then we add +1 to the counter of months with exceeded expenses.

At the end of each pass through the cycle, we add 1 to the pass counter (updating the number of months).

After exiting the loop, we will print the number of months in which Vasya did not have enough money.

```python
month = 0 # Number of months to calculate expenses
count = 0 # Number of months when Vasya spent more than 20,000 rubles

while month < 12:         # While the number of months is less than 12 -
    money = int(input())  # enter the amount
    if money > 20000:     # If the amount is bigger than 20 000
        count += 1        # the number of months when Vasya spent more than 20,000,
                          # becomes 1 more
    month += 1            # We add 1 to the counter, and if we haven't passed yet
                          # all 12 months, then we go through the cycle again

print(count)  # We print at the end the number of months when Vasya would not have enough money
```

#### Final answer

```python
month = 0
count = 0

while month < 12:
    money = int(input())
    if money > 20000:
        count += 1
    month += 1

print(count)
```

## 3. New album
The group "Crimson Fantomas" is going to record an album on a vinyl record, but the record will not contain material longer than 45 minutes. Help the band determine how many songs they can release.

#### INPUT FORMAT
* On each line, enter the length of the song in minutes and seconds in m:ss format.
* Each song is guaranteed to be no longer than 9 minutes 59 seconds.
* Entry ends when the total duration of songs exceeds 45 minutes

#### OUTPUT FORMAT
* Number of songs that will fit on the record.

#### Step 1

First, let's figure out how to read the duration of songs. We write it in the `M:SS` format and we know for sure that the number of minutes does not exceed 9 (that is, the minutes will always be expressed as one digit). Let's try to count the duration of one composition. Next we take the element with index `[0]` - these are minutes, we take the elements `[2]` and `[3]` - these are seconds. We convert minutes and seconds into integers so that we can then calculate the duration of the song in seconds.

```python
song = input()                    # Reading the duration of the track
minutes = int(song[0])            # We take minutes, which is the first element
seconds = int(song[2] + song[3])  # We take seconds
print(minutes)
print(seconds)
print(minutes * 60 + seconds)     # Finding the duration of the song in seconds
```

#### Step 2

Now let's move on to writing the cycle

We create two variables - the total duration of all tracks and a counter for the number of tracks. <br>

The `while` loop runs until all tracks on the album are longer than 45 minutes - 45 * 60 seconds. <br>

At each step of the cycle, we read the duration of the new track.
We add this to the total duration of all tracks, and increase the track number counter by 1. <br>

The cycle ends when the total duration of the album exceeds 45 minutes (45 * 60 seconds), but we have already added the track that is beyond our scope to the counter, although we will not be able to record it in full. Therefore, at the end we print `count - 1`, this is the number of tracks we can *completely* record on the album.

```python
summ = 0   # Duration of all tracks
count = 0  # Number of tracks

# The loop runs until all tracks are longer than 45 minutes, or 45 * 60 seconds
while summ <= 45 * 60:
    song = input()                    # Reading the duration of the track
    minutes = int(song[0])            # We take minutes, which is the first element
    seconds = int(song[2] + song[3])  # We take seconds
    summ += minutes * 60 + seconds    # Convert everything into seconds and
                                      # add to total duration
    count += 1                        # Add this track to the track counter

print(count - 1)  # We display the number of tracks included in the entire album
```

#### Final answer

```python
summ = 0
count = 0
while summ <= 45 * 60:
    song = input()
    minutes = int(song[0])
    seconds = int(song[2] + song[3])
    summ += minutes * 60 + seconds
    count += 1
print(count - 1)
```

## 4. Spendings on lunches
Vasya followed a different diet every month, and then decided to find out in which month he had the cheapest lunches. He calculated and recorded food costs for the past few months. Calculate the least amount of rubles per month Vasya spent on lunches.

#### INPUT FORMAT

* Non-negative integers, each on a new line.
* Then a negative number is entered to indicate the end of the entry.

#### OUTPUT FORMAT
* A number equal to the smallest number entered (not counting the last number, which marks the end of the entry).

#### Step 1

First, we read the first number - the amount of money that Vasya spent in the first month, which will be the current minimum (this is the only amount of spending known to us!).

Then we create a `while` loop that runs as long as the numbers entered are non-negative. Let's try to check how such a cycle works.

```python
# Enter the amount of the first month
money = int(input())

# The amount is higher at the beginning and there is a minimum, because there is nothing to compare with yet
minimum = money

# The end of the sequence will be a negative number, so
# the loop runs as long as the numbers are non-negative (>=0)
while money >= 0:
    money = int(input()) # We take a new number as input
```

#### Step 2

Now at each step we check the `if` condition: whether the entered amount is less than our current minimum, which is stored in the variable. If it is less, then this amount becomes the new minimum. At the end of the cycle step, we read out the amount for the next month of spending.

After exiting the loop (after a negative number is entered), we print the found minimum of the sequence.

```python
# Enter the amount of the first month
money = int(input())

# The amount is higher at the beginning and there is a minimum, because there is nothing to compare with yet
minimum = money

while money >= 0:
    if money < minimum:   # If the entered number is less than the minimum, then
        minimum = money   # it becomes a minimum
    money = int(input())  # We take a new number as input

print(minimum) # At the end we print the minimum spend amount
```

#### Final answer

```python
money = int(input())
minimum = money
while money >= 0:
    if money < minimum:
        minimum = money
    money = int(input())
print(minimum)
```

## 5. Monthly expenses
Vasya noticed that every month he was getting worse and worse at saving money for a new laptop. He thought that this might happen because he was spending more and more every month. Check whether Vasya really spent more in each next month than in the previous one.

#### INPUT FORMAT
* Non-negative integers, each on a new line, then a negative number indicating the end of the input.

#### OUTPUT FORMAT

* True, if in each next month Vasya spent more than in the previous one (without taking into account the last entered number, which means the end of the entry)
* False otherwise.
*If only zero or one month's expenses were entered, the answer should be True.

#### Step 1

We create a variable `increase`, which by default takes the value `True`. We also create a variable where we will record money for the previous period. We start the `while` loop, and set the interruption of its operation `break`, when the entered amount is negative. Let's check how our loop works:

```python
# By default our answer is True, we assume that Vasya really
# spent more and more every month
increase = True

# We will record money for the previous period in this variable,
# to compare with them
old_money = 0

# The loop runs until we force ourselves out of it
while True:
    money = int(input()) # We receive the amount of spending as input

    if money < 0:  # If the amount is still negative, then this signals about
        break      # end of the sequence, use break to exit the loop
```

#### Step 2

Now we add an `if` condition, which is triggered if the sequence does not increase: the new sum is less than or equal to the sum for the previous period, then our `increase` variable becomes `False`, and, regardless of the number of numbers entered further, it no longer will change.

After exiting the loop, the value of the `increase` variable is `True` or `False`, depending on whether the condition in the loop worked or not.

```python
# By default our answer is True, we assume that Vasya really
# spent more and more every month
increase = True

# We will record money for the previous period in this variable,
# to compare with them
old_money = 0

# The loop runs until we force ourselves out of it
while True:
    money = int(input()) # We receive the amount of spending as input

    if money < 0:  # If the amount is still negative, then this signals about
        break      # end of the sequence, use break to exit the loop

    # If the amount for the current period is not greater than the amount for the previous period,
    # then we change our variable to False, and it will no longer have a chance to become True,
    # that is, the condition can be violated once
    if money <= old_money:
        increase = False
        break      # We exit the loop because the value of the variable will not change anymore

    # We write the amount for the current period into a variable with which we will compare
    # in the next period
    old_money = money

print(increase)  # Output the answer True or False
```

#### Final answer

```python
increase = True
old_money = 0

while True:
    money = int(input())

    if money < 0:
        break

    if money <= old_money:
        increase = False
        break

    old_money = money

print(increase)
```
