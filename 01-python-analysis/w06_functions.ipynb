## 1. ATM
The ATM can only give out 50-ruble bills. Write a function that determines whether an ATM can issue a certain amount.

#### REQUIRED FUNCTION 

+ The function `multiples_50`, which takes one parameter — a positive integer `cash` (the amount that the user wants to withdraw). 
+ If the amount is a multiple of 50, the function returns the string 'ATM can issue this amount', otherwise the string 'ATM cannot issue this amount' is returned.

## 2. Grouping of goods

The internal database of the store "Everything for cats" stores a list of all products in the format "<product name>; <product category>". The store "Everything for cats" wants to launch a website for which you need to group products by category. Write a function that will do this.

#### REQUIRED FUNCTION 
+ The `group_goods` function, which takes as a parameter a list of strings in the format `"<product name>; <product category>"'.
+ The function should return a dictionary where the keys are product categories and the values are lists of product names in the corresponding categories.

## 3. Weather station
The staff of the weather station collects a list of weather observations for the day and wants to calculate the minimum, maximum, average temperature values and amplitude. The temperature amplitude is the difference between the maximum and minimum temperature values for the day. Write a function that will help the staff of the weather station. 

#### REQUIRED FUNCTION 
+ The `weather_stat` function, which takes as a parameter a list of real numbers `observations'. 
+ The function should return the minimum, maximum, average values in the list and the amplitude.

## 4. Burger Designer
Create a burger constructor.

#### ТРЕБУЕМЫЕ ФУНКЦИИ 

+ The `ingredients` function, which as a parameter gets a list of `food` containing all the names of the ingredients that should be in the burger filling. The function should display the name of each ingredient on a new line with an indentation of 4 spaces from the beginning of the line. 
+ The `burger` function, which accepts a list of 'food` ingredients as input. The function should first output the string `"(-----bread----)"`, then call the function `ingredients` with the list of `food`, then at the end print the string `"(-----bread-----)"` again.
Note that both functions should not return anything, the burger should be output by calling the `print()` function inside the implemented functions.

## 5. Birthday on the weekend
Ada wants to determine if her birthday falls next year on a day off (Saturday or Sunday). Write a function that will help her check it. 

To solve the problem, use the `weekday` function of the `calendar` module, which determines the day of the week by date. The `weekday` function takes three integers as a parameter: year, month and day, respectively. It returns the number of the day of the week (counting starts from 0 and Monday).

#### REQUIRED FUNCTION 
+ The `is_birthday_weekend` function, which takes as parameters the date of birth (a string in the format 'dd-mm-yyyy') and the year for which the check should be performed (an integer). 
+ If a birthday falls on a Saturday or Sunday in the specified year, the function should return the string 'Hooray, birthday day off!', otherwise, the string 'I'll have to work on my birthday.'
